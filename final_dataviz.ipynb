```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(color_codes=True)
```

```python
cwur = pd.read_csv('cwurData.csv')
cwur
```

```python
title = "Frequencies of Highest World-Rankings for Each Country"
sns.distplot(cwur.groupby('country').min()['world_rank'], kde=False, rug=True).set_title(title);
```

```python
# cwur.groupby('country').sum()['citations']
```

```python
top_10 = cwur.groupby('country').sum().sort_values('citations', ascending=False).head(10).index.values
top_10_df = cwur[cwur.country.isin(top_10)]
# top_10
top_10_df
```

```python
title = "Boxplot of Number of Citations Per University for the Top 10 Countries"
ax = sns.boxplot(x='country', y="citations", data=top_10_df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=50)
ax.set_title(title)
```

```python
title = "Boxplot of Number of Publications Per University for the Top 10 Countries"
ax = sns.boxplot(x='country', y="publications", data=top_10_df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=50)
ax.set_title(title)
```

```python
title = "Relationships Between World Rank and National Rank for the Top 10 Cited Countries"
ax = sns.scatterplot(x="national_rank", y="world_rank", hue='country', data=top_10_df)
ax.set_title(title)
```
